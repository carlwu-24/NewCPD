```python
#Importing libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from datetime import date
from datetime import timedelta
```

```python
website = "https://incidentreports.uchicago.edu/incidentReportArchive.php?startDate="


end = date.today()
end_day = end.day
end_month = end.month
end_year = end.year

start = end - timedelta(days = 6)
start_day = start.day
start_month = start.month
start_year = start.year


website += '{:02d}'.format(start_month) + "%2F" + '{:02d}'.format(start_day) + "%2F" + str(start_year)

website += "&endDate=" + '{:02d}'.format(end_month) + "%2F" + '{:02d}'.format(end_day) + "%2F" + str(end_year)

website += "&offset="

print(website)
```

```python
url = website + "0"
page = requests.get(url)

soup = bs(page.content, "html.parser")
```

```python
try1 = soup.find('li', {"class": "page-count"})
print("First try: ", try1)  # There we go!

try2 = try1.getText()
print("\nSecond try: ", try2)

try3 = try2.split(" / ", 1)
print("\nThird try: " + str(try3[1]))

end = int(try3[1])
```

```python
df = pd.read_html(url)[0]

for i in range(1, end):
  offset = i * 5
  cur = website + str(offset)
  newdata = pd.read_html(cur)[0]
  df = df.append(newdata)

df = df.reset_index().drop(["index",],axis=1)
df
```
